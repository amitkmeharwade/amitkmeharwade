# Query for Ranking Stations by long they have been online

```python
uk_climate1 = sqlContext.sql("SELECT stationname,count(1) as count ,(CASE WHEN count(1) IS NOT NULL THEN DENSE_RANK() over (ORDER BY count(1) DESC NULLS LAST) ELSE NULL END) as station_rank_online  from uk_weather group by stationname")
uk_climate1.show(100)
```

# Query for Ranking Stations by Rainfall, Sunshine

```python
uk_climate2 = sqlContext.sql("SELECT stationname,SUM(rain) as total_rains_in_mm ,(CASE WHEN SUM(rain) IS NOT NULL THEN DENSE_RANK() over (ORDER BY SUM(rain) DESC NULLS LAST ) ELSE NULL END) AS station_rank_by_rainfall ,SUM(sunshine) as total_sunshine_in_hrs,(CASE WHEN SUM(sunshine) IS NOT NULL THEN DENSE_RANK() over (ORDER BY SUM(sunshine) DESC NULLS LAST) ELSE NULL END) as station_rank_by_sunshine from uk_weather group by stationname")
uk_climate2.show(100)
```

# Query for knowing when was the worse rainfall for each station

```python
uk_climate8=sqlContext.sql("select X.stationname,X.year,X.month,X.rain as worse_rainfall_in_mm from uk_weather X INNER JOIN (select distinct a.stationname,MAX(a.rain) as max_rain from uk_weather a inner join uk_weather b where a.stationname=b.stationname and a.year=b.year and a.month=b.month group by a.stationname)Y where TRIM(X.stationname)=TRIM(Y.stationname) and TRIM(X.rain)=TRIM(Y.max_rain) ORDER BY worse_rainfall_in_mm DESC ")

uk_climate8.show(100)
```

# Query for knowing when was the best sunshine for each station

```python
uk_climate9=sqlContext.sql("select X.stationname,X.year,X.month,X.sunshine as best_sunshine_in_hrs from uk_weather X INNER JOIN (select distinct a.stationname,MAX(a.sunshine) as max_sunshine from uk_weather a inner join uk_weather b where a.stationname=b.stationname and a.year=b.year and a.month=b.month group by a.stationname)Y where TRIM(X.stationname)=TRIM(Y.stationname) and TRIM(X.sunshine)=TRIM(Y.max_sunshine) ORDER BY best_sunshine_in_hrs DESC ")

uk_climate9.show(100)
```

# Query for average for MAY across all stations

```python
uk_climate11=sqlContext.sql("select distinct a.month,a.year,AVG(a.tmax) as avg_max_temperature,AVG(a.tmin) as avg_min_temperature,AVG(a.airfrost) as avg_air_frost,AVG(a.rain) as avg_rainfall,AVG(a.sunshine) as avg_sunshine from uk_weather a cross join uk_weather b where a.month=5 and b.month=5 group by a.month,a.year order by year desc")
uk_climate11.show(200)
```
